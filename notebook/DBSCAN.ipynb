# IMPLEMENTACIÓN DE DBSCAN SIN LIBRERÍAS

```python
from sklearn.neighbors import KDTree


def db_scan( r, umbral):
   etiqueta = -1
   labels = np.zeros(x.shape[0])
   for index in range(x.shape[0]):
     if(labels[index] != 0):
       continue
     tree = KDTree(x,metric='euclidean')  
     ind = tree.query_radius(x[index].reshape(-1,31), r=r) ## analizar
     if(len(ind[0])<umbral): ## analizar si es len o no
       labels[index] = -1
       continue
     etiqueta = etiqueta + 1
     labels[index] = etiqueta
     arr = ind[0]

     for index_ in arr:
        if(index_ ==index or labels[index_] > 0):
            continue
        if(labels[index_] == -1):
            labels[index_] = etiqueta
        ind_ = tree.query_radius(x[index_].reshape(-1,31), r=r) ## analizar
        labels[index_] = etiqueta
        if(len(ind_[0]) < umbral):
           continue
       
        arr = np.append(arr,ind_[0])
    
   return labels
       

a = db_scan(132, 3)
a = a.astype(int)
print(a)
```
